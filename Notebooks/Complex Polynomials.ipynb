```python
from bqplot import *
from bqplot import pyplot as plt
import matplotlib.patches as patches
from IPython.display import display
```

```python
x_data = range(10)
y_data = [i ** 2 for i in x_data]

x_sc = LinearScale()
y_sc = LinearScale()

ax_x = Axis(label='Test X', scale=x_sc, tick_format='0.0f')
ax_y = Axis(label='Test Y', scale=y_sc, orientation='vertical', tick_format='0.2f')

line = Lines(x=x_data,
             y=y_data,
             scales={'x': x_sc, 'y': y_sc},
             colors=['green'],
             stroke_width=30)

scatter = Scatter(x=x_data,
             y=y_data,
             scales={'x': x_sc, 'y': y_sc},
             colors=['green'],
             stroke_width=30,marker="circle")

def clickEvnt(c,y):
    print("dd")
    if(line.colors == ['green']):
        line.colors = ['red']
    else:
        line.colors = ['green']
    
line.on_element_click(clickEvnt, remove=False)

fig = Figure(axes=[ax_x, ax_y], marks=[line,scatter])

display(fig)
```

```python
z = plt.figure()
k = patches.Rectangle( (0,0),20,20,color="red")
a = plt.scatter([100], [44])
a.on_element_click(clickEvnt, remove=False)
#plt.show()
z.add_patch(k)
```
